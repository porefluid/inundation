```python
import os 
import rasterio as rio
import rasterio.plot as plot
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
data_dir = '/Users/corey/Documents/Research/CHART/data'
fig_dir = '/Users/corey/Documents/Research/CHART/figures'
```

```python
results = rio.open(os.path.join(data_dir, 'ex_results', 'example_results.tif'))
min_s1 = rio.open(os.path.join(data_dir, 'ex_results', '2018_min_backscatter.tif'))
max_s1 = rio.open(os.path.join(data_dir, 'ex_results', '2018_max_backscatter.tif'))
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 8), sharey=True)

image_hidden = ax[0].imshow(results.read(1), vmin=0, vmax=200)
plot.show(results.read(1), ax=ax[0], transform=results.transform, vmin=0, vmax=200)

image_hidden2 = ax[1].imshow(results.read(3), vmin=200, vmax=365)
plot.show(results.read(3), ax=ax[1], transform=results.transform, vmin=200, vmax=365)

divider = make_axes_locatable(ax[0])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(image_hidden, cax=cax)

divider2 = make_axes_locatable(ax[1])
cax2 = divider2.append_axes("right", size="5%", pad=0.05)
plt.colorbar(image_hidden2, cax=cax2)
cax2.set_ylabel('Days since May 1, 2018')

for a in ax:
    a.set_ylim(12.37, 12.58)
    a.set_xlim(76.35, 76.6)
ax[0].set_title('First "inundated" observation (VH)')
ax[1].set_title('Last "inundated" observation (VH)')
plt.savefig(os.path.join(fig_dir, 'example_results_2018_vh.png'), dpi=300, facecolor='white')
```
